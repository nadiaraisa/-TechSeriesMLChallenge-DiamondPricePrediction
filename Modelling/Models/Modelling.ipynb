```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings
warnings.filterwarnings("ignore")

from pycaret.regression import *
```

```python
train_path = '../../Preprocessing/Preprocessed/full_train.csv'
test_path = '../../Preprocessing/Preprocessed/full_test.csv'

train = pd.read_csv(train_path, index_col='Unnamed: 0')
test = pd.read_csv(test_path, index_col='Unnamed: 0')
```

```python
full_data = pd.concat([train, test], axis=0).reset_index(drop=True)
```

## Using Pycaret for Regression

```python
s = setup(data=full_data, target='price')
```

```python
compare_models() # catboost regressor has the best results
```

```python
cb = create_model('catboost')
```

## Export Model

```python
import pickle

file = open('catboost_regressor.pkl', 'wb')
pickle.dump(cb, file)
file.close()
```
